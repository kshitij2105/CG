```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
df.info
```

```python
df['Species'].value_counts()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
KS = {'color': ['cyan', 'pink', 'purple']}
sns.FacetGrid(df,hue='Species',hue_kws=KS)\
  .map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
  .add_legend()
```

```python
sns.boxplot(x="Species",y="SepalWidthCm",data=df)
```

```python
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=df,jitter=True,linewidth=2,edgecolor='pink')
```

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=df,size=6,color="purple")
```

```python
sns.swarmplot(x='Species',y='SepalLengthCm', data=df, palette="Set2")
```

```python
sns.pairplot(df,hue='Species')
```

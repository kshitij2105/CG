```python
df.sort_values(by=["Post Hour"],inplace=True)
df

#By executing df.sort_values(by=["Post Hour"], inplace=True), the DataFrame df will be sorted in ascending 
#order based on the values in the "Post Hour" column. 
```

```python
df_transpose = df.T
df_transpose
```

```python
pivot_table = pd.pivot_table(df,index = ['Type','Category'],values = 'comment')
print(pivot_table)

#By calling pd.pivot_table(df, index=['Type', 'Category'], values='comment'), you create a pivot table
#with the 'Type' and 'Category' columns as the index, and the 'comment' column as the values. The pivot 
#table aggregates the 'comment' values based on the combinations of 'Type' and 'Category' present in the DataFrame.
```

```python
melt_df = pd.melt(df,id_vars=['Page total likes'],value_vars=['Category','Post Month','Post Weekday'])
melt_df
```

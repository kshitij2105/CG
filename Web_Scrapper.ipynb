```python
import requests #this library is used to send HTTP request to a specified URL using get() post() put() methods 
#here we are using only get() method
#Inshort we are either requesting data from a website or we are pushing the data to a webserver
#inour case we are just retriving the data from a website
#in short it works as a request and response protocol between a client and a server
import bs4
#bs4 is used for pulling data out og html and xml files
```

```python
request1 = requests.get('https://www.flipkart.com/realme-x7-pro-5g-mystic-black-128-gb/p/itmd2848609bbe77?pid=MOBFZHDZDVKFXQZH&lid=LSTMOBFZHDZDVKFXQZHYCD0RV&marketplace=FLIPKART&q=realme+x7+pro&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_2_9_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_9_na_na_na&fm=organic&iid=05ba8ae7-4698-4760-9b06-d1bb976aaf8a.MOBFZHDZDVKFXQZH.SEARCH&ppt=hp&ppn=homepage&ssid=4f7xf5b7ao0000001685196415277&qH=741ae102d223a7dc')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
#we have to store this retrived data into a variable called soup for that we are using bs4 library
#for that we are using BeautifulSoup function  It provides a convenient way to extract data from web pages by
#traversing the document's structure and locating specific elements or tags.
```

```python
reviews = soup.find_all('div',{'class': 't-ZTKy'});
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
#Average overall rating
ratings = soup.find('div',{'class':'_2d4LTz'}).get_text();
print(ratings)
```

```python
#individual Ratings
individual_ratings = soup.findAll('div',{'class':'_3LWZlK _1BLPMq'})
for indi_rating in individual_ratings :
    print(indi_rating.get_text() + " \n")
```

```python
tags = soup.find('span',{'class':''}).get_text();
tags
```

```python
#Fetching customer name
customer_name = soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
for customer_name in customer_name :
    print(customer_name.get_text() + " \n")
```

```python
#Fetch price of product
price = soup.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text();
print(price)
```

```python
questions_div = soup.findAll('div', {'class': '_1xR0kG _3cziW5'})
for question in questions_div:
    print(question.get_text())
    answer = question.findNext('div', {'class': '_2yeNfb'}).get_text()
    print(answer + "\n\n")

```
